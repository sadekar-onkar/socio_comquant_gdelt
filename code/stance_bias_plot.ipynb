# Stance Scores per Bias Category

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
articles = pd.read_csv("data/articles_final_wStance.csv")

articles = articles[articles["bias_rating"] != "LEAST BIASED"].reset_index(drop=True)
```

```python
def bias_rename(name):
    if name == "LEFT-CENTER (by Saudi standards)":
        return "LEFT-CENTER"
    elif name == "LEFT CENTER":
        return "LEFT-CENTER"
    elif name == "LEFT BIASED":
        return "LEFT"
    elif name == "FAR LEFT":
        return "FAR-LEFT"
    else:
        return name
```

```python
articles["bias_rating"] = articles["bias_rating"].apply(lambda x: bias_rename(x))
```

```python
plt.figure(figsize=(16, 12))
sns.violinplot(x="bias_rating", y="stance_score", data=articles,
               order=["FAR-LEFT", "LEFT", "LEFT-CENTER", "RIGHT-CENTER", "RIGHT"], linewidth=3)
plt.title("Stance Scores per Bias Category", fontsize=22)
plt.xlabel("Bias Category", fontsize=16)
plt.ylabel("Stance Score", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.savefig('figures/violin_plot.png', bbox_inches='tight')
```
